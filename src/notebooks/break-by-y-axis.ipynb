```python

interval = 1.0
extractor = DisplacementIntervalSequenceExtractor(
    interval = interval,
    tolerance = 0.1,
    yLow = pedDf["sceneY"].min(),
    yHigh = pedDf["sceneY"].max(),
    xCol="sceneX",
    yCol="sceneY"
)

repo = IntervalPatternRepository(intervalType="distance")
```

```python
pedIds = northenedDf["uniqueTrackId"].unique()
for pedId in tqdm(pedIds):
    sourceId = str(pedId)
    aTrack = northenedDf[northenedDf["uniqueTrackId"] == pedId]
    patterns = extractor.extract(sourceId, aTrack)
    repo.addSequence(
        sourceId=sourceId, interval=interval,
        patterns=patterns
    )
    
```

```python
pedDf.head()
```

```python
minYDisplacement = 3
maxXDisplacement = 6

visualizer = TrajectoryVisualizer()
colMapper = ColMapper(
        idCol='trackId', 
        xCol='mapX', 
        yCol='mapY',
        xVelCol='sceneXVelocity', 
        yVelCol='sceneYVelocity', 
        speedCol='speed',
        fps = 25.0
    )
transformer = TrajectoryTransformer(colMapper)
cleaner = TrajectoryCleaner(
    colMapper = colMapper,
    minSpeed = 0.0, # we can consider them as obstacles
    maxSpeed = 3.5,
    minYDisplacement = minYDisplacement,
    maxXDisplacement = maxXDisplacement
)
```

```python
transformer.deriveAxisVelocities(pedDf)
transformer.deriveSpeed(pedDf)
transformer.deriveDisplacements(pedDf)
pedDf.describe()
```

```python
cleanedPedDf = northenedDf[northenedDf["speed"] < 5].copy()
assert len(cleanedPedDf["trackId"].unique() == northenedDf["trackId"].unique())
```

```python
interval = 1.0
extractor = DisplacementIntervalSequenceExtractor(
    interval = interval,
    tolerance = 0.1,
    yLow = cleanedPedDf["mapY"].min(),
    yHigh = cleanedPedDf["mapY"].max(),
    xCol="mapX",
    yCol="mapY"
)

repo = IntervalPatternRepository(intervalType="distance")

pedIds = cleanedPedDf["trackId"].unique()
sourceTrajs = []
for pedId in tqdm(pedIds):
    sourceId = str(pedId)
    aTrack = cleanedPedDf[cleanedPedDf["trackId"] == pedId]
    patterns = extractor.extract(sourceId, aTrack)
    if(len(patterns) == 0):
        print(f"0 patterns for track {pedId}")
    else:
        sourceTrajs.append(baseGenerator.combinePatterns(patterns))
        repo.addSequence(
            sourceId=sourceId, interval=interval,
            patterns=patterns
        )
```
